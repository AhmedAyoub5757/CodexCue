```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Suicide_bombing_attacks.csv')
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from math import pi
```

```python
# Bubble Map: Locations of Blasts with Casualties
# Convert columns to numeric and handle non-numeric values
df['Killed Max'] = pd.to_numeric(df['Killed Max'], errors='coerce').fillna(0)
df['Injured Max'] = pd.to_numeric(df['Injured Max'], errors='coerce').fillna(0)

# Calculate total casualties for each blast
df['total_casualties'] = df['Killed Max'] + df['Injured Max']

fig = px.scatter_geo(df,
                     lat='Latitude',
                     lon='Longitude',
                     size='total_casualties',
                     color='Province',
                     hover_name='City',
                     size_max=25,
                     title='Locations of Blasts with Casualties')
# Set the geographical scope to Asia and center on Pakistan
fig.update_geos(scope='asia', 
                center={'lat': 30.3753, 'lon': 69.3451}, 
                projection_scale=5)  # Adjust projection_scale as needed for zoom level
fig.show()
```

```python
import plotly.express as px

# Aggregate the data by city
city_counts = df['City'].value_counts().reset_index()
city_counts.columns = ['City', 'count']

# Create the Bar Chart
fig = px.bar(city_counts, 
             x='City', 
             y='count', 
             title='Number of Blasts by City', 
             labels={'count': 'Number of Blasts', 'city': 'City'})
fig.show()
```
